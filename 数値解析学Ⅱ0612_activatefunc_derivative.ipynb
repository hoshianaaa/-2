```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python

def sigmoid(a):
    return 1 / (1 + np.exp(-a))

def d_sigmoid(a):
    return sigmoid(a)*(1-sigmoid(a))
```

```python

x = np.arange(-10, 10, 0.1)

y = sigmoid(x)
dydx = d_sigmoid(x)


plt.plot(x, y)
plt.plot(x, dydx)
plt.show()
```

```python

def d_tanh(a):
    return 4 / pow((np.exp(a)+np.exp(-a)) ,2)
```

```python

x = np.arange(-10, 10, 0.1)

y = np.tanh(x)
dydx = d_tanh(x)


plt.plot(x, y)
plt.plot(x, dydx)
plt.show()
```

```python

def relu(x):
	return max(0.0, x)

def d_relu(x):
    if x < 0:
        return 0
    else:
        return 1
```

```python

x = np.arange(-2, 2, 0.01)
y = [relu(i) for i in x]
dydx = [d_relu(i) for i in x]

plt.plot(x, y)
plt.plot(x, dydx)
plt.show()

```
